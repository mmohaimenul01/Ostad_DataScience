**Find the model Score**

```python
#Code here please, so that you dont lose the output
df = pd.read_csv("Aust_Credit_Approval_Data.csv")
X = df.drop(columns=["target"])
y = df["target"].astype(int)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=1, shuffle=True 
)
svm_model = SVC(C=20, gamma="auto", kernel="linear")
svm_model.fit(X_train, y_train)
print("Train score (accuracy):", svm_model.score(X_train, y_train))
print("Test score  (accuracy):", svm_model.score(X_test, y_test))
```
